```python
np.random.seed(34)
marks = np.random.randint(0,101,size=[6,7])
```

```python
mean_marks = marks.mean() 
```

```python
idx,idy = np.where(marks >= mean_marks) #it returns row and column index
```

```python
marks[idx,idy] #it returns the values of marks which are greater than mean_marks
```

```python
marks = np.where(marks >= mean_marks, "Pass","Fail")
marks
```
