```python
import numpy as np
```

```python
# ================================ index and slicing =============================
```

```python
np.random.seed(45)
random = np.random.randint(0, 225, size=[5,6])
random
```

```python
random[1:4:2, 1:5:2]
```

```python
random[1,1]
```

```python
np.random.seed(45)
random = np.random.randint(0, 225, size=[5,6])
random
```

```python
random[1:4,1::2]
```

```python
random[1:4,:5:2]
```

```python
# reversed slicing

np.random.seed(45)
random = np.random.randint(0, 225, size=[5,6])
random
```

```python
random[3:0:-1,3:0:-1]
```

```python
np.random.seed(45)
random = np.random.randint(0, 225, size=[5,6])
random
```

```python
random[1:4,0:3]
```

```python
# revesed with negative indexing

random[-2:-5:-1,-4::-1]
```

```python
# concatenation 

arr1 = random[0:3,0:3]
arr1
```

```python
arr2 = random[2:5,2:5]
arr2
```

```python
arr = np.concatenate((arr1, arr2), axis=0)
arr
```
