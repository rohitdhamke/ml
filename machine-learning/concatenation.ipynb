```python
import numpy as np
```

```python
# concatenation and stacking
```

```python
# joining two or more arrays together making them a array is known as concatenation. The numpy.concatenate() function is used to join two or more arrays of the same shape along a specified axis.
```

```python
arr1 = np.array([2,4,5,6])
arr2 = np.array([1,3,7,8])

arr = np.concatenate([arr1, arr2])
```

```python
arr
```

```python
arr = np.concatenate([arr1, arr2])
# arr = np.concatenate([arr1, arr2], axis=1)
```

```python
# ======  type of stacking  =======
#   1. Horizontal stacking: np.hstack()
#   2. Vertical stacking: np.vstack()
```

```python
np.random.seed(40)
arr1=np.random.randint(0,50,size=[2,4])
print(arr1)
print("\n")
arr2=np.random.randint(0,100,size=[2,4])
print(arr2)

```

```python
arr = np.hstack([arr1, arr2])
```

```python
arr
```

```python
arr=np.vstack([arr1, arr2])
arr
```

```python
np.random.seed(34)
# create a 2 random arrays of range 0-255 and shape 2,2 
# and one of range 0-344 and shape 2,2
# join them such that the elemnt of the first array 2d arrays are in concatenation with the second arrays
# and also convert into 1d array
```

```python
arr1 =np.random.randint(0,255,size=[2,2])
print(arr1)
print(arr1.shape)
print("\n")
arr2 =np.random.randint(0,344,size=[2,2])
print(arr2)
print(arr2.shape)
```
