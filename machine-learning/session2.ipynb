```python
import numpy as np
```

```python
np.random.seed(67)
random = np.random.randint(0,255, size=[4,4])
```

```python
random
```

```python
(2,3)
```

```python
rows , cols = random.shape
```

```python
#creating strides of 2x2
for i in range(rows-1):
    for j in range(cols-1):
        print(random[i:i+2, j:j+2])
```
