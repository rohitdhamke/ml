```python
import pandas as pd
```

```python
#             ======================== group-by and aggregation ==========================
```

```python
data = pd.read_csv(r"/workspaces/ml/panda/sample_-_superstore (1).csv",encoding="latin-1")
```

```python
data
```

```python
new_data = data.groupby("Customer Name").agg(Total_Sales = ("Sales", "sum")).reset_index()
```

```python
new_data
```

```python
# aggregate function ==> mean, max, min, count, sum 

new_data = data.groupby("Customer Name").agg(Total_Sales = ("Sales", "sum"), Avg_sal = ("Sales", "mean")).reset_index()
new_data
```

```python
customer_arriaval = data.groupby("Customer Name").agg(Arriaval = ("Customer Name", "count")).reset_index()
customer_arriaval
```

```python
#               ====================== sort =============================
```

```python
customer_arriaval.sort_values("Arriaval", ascending= False)
```

```python
# ==================  merging two dataset  ============================
```

```python
data1 = pd.read_csv(r"/workspaces/ml/panda/Product_ID,Product_Name,Profit_2014.csv")
data2 = pd.read_csv(r"/workspaces/ml/panda/Product_ID,Product_Name,Profit_2015.csv")
```

```python
data1
data2
print(data1)
print("\n")
print(data2)
```

```python
data1.isnull().sum()
```

```python
#  ============   dropna method   ====================
#  for removing null values
```

```python
data1 = data1.dropna()
```

```python
data1
```

```python
mean_profit = data1["Profit_2014"].mean()
```

```python
mean_profit
```

```python
data1 = data1.fillna(mean_profit)
```

```python
data1
```

```python
data2.isnull().sum()
```

```python
data2 = data2.fillna(data2["Profit_2015"].mean())
```

```python
data2
```

```python
datamerg = pd.merge(data1,data2, how="inner", on="Product_Name")

# inner
# left
# right
# outer
```

```python
print(data1)
print("\n")
print(data2)
print("\n")
datamerg
```
