```python
# create data frame

diction = {"Name" : ["rohit","manas","ninad","raj"], "Salary" : [10000,20000,30000,40000], "Performance":[4,7,2,6]}
```

```python
newdata=pd.DataFrame(diction, index=[101,102,103,104])
newdata
```

```python
# =========================================== iloc ==================================================
```

```python
# iloc = iloc is indexing based locator through which we can access a particular set of values.


newdata.iloc[1,0]
```

```python
# newdata.iloc[0:3,0:]
newdata.iloc[0:3,0:2]
```

```python
# =========================================== loc ==================================================
```

```python
# loc : loc is labled based indexing and in loc end value is not exclusive
```

```python
newdata.loc[101:102,"Name":"Performance":2]
# newdata.loc[101:102,"Name":"Performance"]
```
